```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv("iterations_timing.csv")

# Step 2: Extract Data
iterations = df["Iterations"]  # Assuming "Iterations" is the column name for iteration values

# Assuming "Vectorised Time", "Numba Time", ..., "Unedited code Time" are the column names for timing data
method_columns = {
    "Vectorised Time": "vectorization",
    "Numba Time": "Numba",
    "Numba vectorisation Time": "Numba Vectorization",
    "Cython Time": "Cython",
    "MPI Time": "OpenMPI",
    "Unedited code Time": "Unedited code"
}

# Step 3: Plotting
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

# Plot timing vs. iteration for each method
for column_name, label in method_columns.items():
    plt.plot(iterations, df[column_name], label=label)

# Add labels and title
plt.xlabel("Iteration")
plt.ylabel("Timing (ms)")  # Adjust units if needed
plt.title("Timing vs. Iteration for Different Methods")

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Directory containing the text files
directory = r"C:\Users\Nischayee\AC1 and AC2 assessment"

# List of file names with raw string literals
file_names = [
    r"LL-Output-Fri-15-Mar-2024-at-06-10-14AM.txt",
    r"LL-Output-Fri-15-Mar-2024-at-06-19-08AM.txt",
    r"LL-Output-Fri-15-Mar-2024-at-06-20-23AM.txt",
    r"LL-Output-Fri-15-Mar-2024-at-06-21-55AM.txt",
    r"LL-Output-Fri-15-Mar-2024-at-06-22-53AM.txt",
    r"LL-Output-Fri-15-Mar-2024-at-06-23-48AM.txt"
]

# Custom labels for legend
custom_labels = [
    "Original", "Vectorization", "Numba", "Numba Vectorization", "Cython", "MPI"
]

# Function to extract iteration and order parameter data from a file
def extract_data(file_path):    iterations = []
    order_parameters = []
    with open(file_path, 'r') as file:
        for line in file:
            # Skip lines that don't contain numeric data or comments
            if line.startswith("#") or not line.strip():
                continue
            # Split the line into its individual values
            values = line.strip().split()
            # Extract the iteration and order parameter values
            iteration = int(values[0])
            order_parameter = float(values[3])
            # Append to the lists
            iterations.append(iteration)
            order_parameters.append(order_parameter)
    return iterations, order_parameters

# Plot iteration vs. order parameter for each file
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
for file_name, label in zip(file_names, custom_labels):
    file_path = os.path.join(directory, file_name)
    iterations, order_parameters = extract_data(file_path)
    plt.scatter(iterations, order_parameters, label=label)

# Add labels and title
plt.xlabel("Iteration")
plt.ylabel("Order Parameter")
plt.title("Iteration vs. Order Parameter")
plt.legend()  # Show legend with custom labels
plt.grid(True)  # Add grid
plt.show()
```
